```python
#importing the libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#importing the dataset
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv('/content/drive/My Drive/datasets/Salary_Data.csv')
```

```python
X = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
#Splitting the datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
X_train = X_train.reshape(-1,1)
regressor.fit(X_train, y_train)
```

```python
X_test = X_test.reshape(-1,1)
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train, y_train, c = 'red')
plt.plot(X_train, regressor.predict(X_train), c = 'green')
plt.title('Salary vs Experience(Training set)')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test, y_test, c='red')
plt.plot(X_test, y_pred, c= "blue") # since this line is same for test and train we can still use // plt.plot(X_train, regressor.predict(X_train), c = 'green') and it will give same result
plt.title("Predictions")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
